**Actividad 4:** Dado el siguiente array `arr = np.array([10, 20, 30, 40, 50])`
1. Suma 5 a cada elemento.
2. Multiplica todo por un numero aleatorio (utiliza el modulo random para generarlo.)

```python
#todo:

import numpy as np
import random

# crear el array
array = np.array([10, 20, 30, 40, 50])

# sumar 5 a cada elemento
array_sumado = array + 5

# ultiplicar por un número aleatorio 
numero_aleatorio = random.uniform(1, 10)  # genera un número aleatorio entre 1 y 10
array_multiplicado = array_sumado * numero_aleatorio

print("Array original:", array)
print("Array después de sumar 5:", array_sumado)
print("Array después de multiplicar por {numero_aleatorio}:", array_multiplicado)
```

**Actividad 6:** Crea un array que comience en 0 hasta el 20, pero que solo tenga numeros pares y luego realiza las 4 operaciones basicas. (suma, resta, multiplicacion, division).

```python
#todo: 


import numpy as np

array_par = np.arange(0, 21, 2)

#arange crea un array con números desde 0 hasta 20, con un paso de 2 (números pares)

# mostrar el array de números pares
print("array de números pares:", array_par)


# sumar 5 a cada elemento 
suma = array_par + 5

#restar 3 a cada elemento 
resta = array_par - 3

# multiplicar por 2 cada elemento 
multiplicaacion = array_par * 2

# dividir entre 2 cada elemento 
division = array_par / 2

# mostrar los resultados de las operaciones
print("Array con suma de  5:", suma)
print("Array con resta de 3:", resta)
print("Array con multiplicación  por 2:", multiplicaacion)
print("Array  con division entre 2:", division)
```
